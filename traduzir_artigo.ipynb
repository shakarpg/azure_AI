```python
import requests
from bs4 import BeautifulSoup

def extract_text_from_url(url):
    response = requests.get(url)

    if response.status_code == 200:
        soup=BeautifulSoup(response.text,'html.parser')
        for script_or_style in soup(['script','style']):
            script_or_style.decompose()
        texto = soup.get_text(separator=' ')
        #Limpar texto
        lines = (line.strip() for line in texto.splitlines())
        chunks = (phrase.strip() for line in lines for phrase in line.split("  "))
        texto_limpo = '\n'.join(chunk for chunk in chunks if chunk)
        return texto_limpo
    else:
        print(f"Failed to retrieve the webpage. Status code: {response.status_code}")
        return None

url = ('https://dev.to/kenakamu/azure-open-ai-in-vnet-3alo')
text = extract_text_from_url(url)
print(text)
```

```python
from langchain_openai.chat_models.azure import AzureChatOpenAI
from langchain_openai.embeddings.azure import AzureOpenAIEmbeddings
from google.colab import userdata


client = AzureChatOpenAI(
    azure_endpoint= "https://oia-dio-bootcamp-dev-eastus-002.openai.azure.com/",
    api_key= userdata.get('TRADUTOR_AZURE'),
    api_version="2025-01-01-preview",
    deployment_name="gpt-4o-mini",
    max_retries=0
    )

def translate_article(text,lang):
  messages = [
      ("system", "Você atua como tradutor de textos"),
      ("human", f"Traduza o {text} para o idioma {lang} e responda em markdown")
  ]

  response = client.invoke(messages)
  print(response.content)
  return response.content

translate_article("Even though the models themselves are the same, there are many differences including", "português")
```
